```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4],[12,43,25,15])
plt.show()
```

```python
plt.title('plotting')
plt.plot([10,20,30,40])
plt.show()
```

```python

plt.title('legend')
plt.plot([10,20,30,40], label='asc',color='r')
plt.plot([40,30,20,10], label='desc',color='b')
plt.legend()
plt.show()
```

```python

plt.title('legend')
plt.plot([10,20,30,40], label='dashed',color='r', linestyle='--')
plt.plot([40,30,20,10], label='dotted',color='b', ls=':')
plt.legend()
plt.show()
```

```python
plt.title('simplified one')
plt.plot([10,20,30,40],'r--')
plt.plot([23,43,56,21],'p:')
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40],'r.',label='circle')
plt.plot([23,43,56,21],'g^',label='triangle up')
plt.plot([33,44,55,66],'b.--') # ,,  
plt.legend()
plt.show()
```
