```python
f=open('seoul.csv')
data = csv.reader(f)
header=next(data)
aug=[]
jan=[]

for row in data:
    month = row[2].split('-')[1]
    if row[4] !='':
        if month =='08':
            aug.append(float(row[4]))
        elif month=='01':
            jan.append(float(row[4]))
            

plt.hist(aug,bins=100,color='red',label='Aug')
plt.hist(jan,bins=100,color='blue',label='Jan')
plt.legend()
plt.show()
```

```python
# 8월 일별 기온데이터
import matplotlib.pyplot as plt
import csv

f=open('seoul.csv')
data=csv.reader(f)
next(data)

day=[[] for _ in range(31)]

for row in data:
    if row[4] != '' :
        date = row[2].split('-')
        if date[1]=='08':
            day[int(row[2].split('-')[2])-1].append(float(row[4]))

plt.style.use('ggplot') # graph style
plt.figure(figsize=(10,5),dpi=300) # graph size
plt.boxplot(day, showfliers=False) # do not show outliers
plt.show()
```
