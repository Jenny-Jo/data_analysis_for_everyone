```python
import csv
f=open('seoul.csv')
data = csv.reader(f)
header=next(data)

result=[]

for row in data:
    if row[4] != '' : # 최저기온
        result.append(float(row[4]))

print(result)
f.close()
```

```python
import matplotlib.pyplot as plt
plt.plot(result,'r')
plt.figure(figsize=(10,2)) # 그래프 크기 가로,세로 설정
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data = csv.reader(f)
header = next(data)

high=[]
low=[]

# for i in data:
#     print(i)
for row in data:
    if row[4] !='' and row[-3] != '':
        date=row[2].split('-')
        if 2005 <= int(date[0]):
            if date[1]=='05' and date[2]=='09':
                high.append(float(row[4]))
                low.append(float(row[-3]))
plt.plot(high,'pink',label='high')        
plt.plot(low,'blue',label='low')
plt.legend()

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('Graph of temperature change on my birthdays')

plt.show()
f.close()
```
